```python
Date_news = {}
import string
import nltk

stopwords = nltk.corpus.stopwords.words('english')
p = string.punctuation
d = string.digits
table_p = string.maketrans(p, len(p) * " ")
table_d = string.maketrans(d, len(d) * " ")
with open('RedditNews.csv','r') as f:
    content = f.read()
    i = 1
    for i in content.splitlines():
        try:
            news = i.split(',')
            #print(news[1])
            s = news[1].translate(table_d)
            s = s.lower().translate(table_p)
            #print(s)
            w_list = [word for word in s.split() if word not in stopwords and len(word)>2]
            s = " ".join(w_list)
            #print('\n')
            if news[0] not in Date_news:
                Date_news[news[0]] = s
            else:
                Date_news[news[0]] += ' ' + s
        except:
            continue
```

```python
date_headlines_df = pd.DataFrame(Date_news.items(), columns=['Date', 'Headlines'])
```

```python
from textblob import TextBlob
subjectivity = []
polarity = []

for i in date_headlines_df['Headlines']:
    tb_sent = TextBlob(i)
    subjectivity.append(tb_sent.subjectivity)
    polarity.append(tb_sent.polarity)
    #anger.append(tb_sent.classify)
```

```python
date_headlines_df['Subjectivity'] = subjectivity
date_headlines_df['Polarity'] = polarity
```

```python
date_headlines_df.to_csv('Cleaned_Data.csv')
```

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
Headlines_stem = []
for sen in date_headlines_df['Headlines']:
    t = []
    for word in sen.split():
        #print word
        t.append(porter_stemmer.stem(word))
    #print t
    s = ' '.join(t)
    Headlines_stem.append(s)
    #print s
```

```python
date_headlines_df['Stemmed_Headline'] = Headlines_stem
```
